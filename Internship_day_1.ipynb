```
import pandas as pd
df=pd.read_csv("/content/titanic uncleaned.csv")
df.head()
```

```
df.isnull().sum()
```

```
df
```

```
df['age_mean_imputed']=df['Age'].fillna(df['Age'].mean())
df.head()
```

```
df['age_median_imputed']=df['Age'].fillna(df['Age'].median())
```

```
df.head()
```

```
df['age_mode_imputed']=df['Age'].fillna(df['Age'].mode()[0])
df.head()
```

```
df[['Age','age_mean_imputed']]
```

```
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
sns.boxplot(data=df,x='Pclass',y='Survived',palette='pastel')




```

```
plt.figure(figsize=(10,6))
plt.title('box plot of age by passenger class',fontsize=15)
plt.bar(df['Pclass'],df['Age'])
plt.xlabel('passenger class',fontsize=15)
plt.ylabel('age',fontsize=15)

plt.show()
```

```
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Embarked',y='Fare',palette='pastel')


```

```
import pandas as pd
from sklearn.preprocessing import LabelEncoder
encoded_data=df.copy()
label_encoder=LabelEncoder()
encoded_data['sex_label']=label_encoder.fit_transform(encoded_data['Sex'])
```

```
one_hot_encoded_data=pd.get_dummies(encoded_data,columns=['Embarked'],prefix=['Embarked'])
encoded_data=pd.concat([encoded_data,one_hot_encoded_data],axis=1)
```
