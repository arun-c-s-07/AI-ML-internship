```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
encoded_data=df.copy()
label_encoder=LabelEncoder()
encoded_data['Gender']=label_encoder.fit_transform(encoded_data['Gender'])
encoded_data.head()
```

```python
department_dummies = pd.get_dummies(encoded_data['Department'], prefix='Department', drop_first=False)
encoded_data = encoded_data.drop(columns=['Department'])
encoded_data = pd.concat([encoded_data, department_dummies], axis=1)
encoded_data.head()
```

```python
p_mapping={
    1:'First',2:'second',3:'third',4:'fourth',5:'five'
}
encoded_data['P_Rating'] = encoded_data['Performance_Rating'].iloc[:, 0].map(p_mapping)
```

```python
print(encoded_data.columns)
```

```python
encoded_data
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
numerical_cloumns=['Age','Experience_Years']
encoded_data.dropna(subset=numerical_cloumns,inplace=True)
encoded_data
```

```python

minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()


encoded_data_unique_cols = encoded_data.loc[:, ~encoded_data.columns.duplicated(keep='first')]

encoded_data['age_minmax'] = minmax_scaler.fit_transform(encoded_data_unique_cols[['Age']])
encoded_data['E_MinMax'] = minmax_scaler.fit_transform(encoded_data_unique_cols[['Experience_Years']])
encoded_data['age_standard'] = standard_scaler.fit_transform(encoded_data_unique_cols[['Age']])
encoded_data['E_standard'] = standard_scaler.fit_transform(encoded_data_unique_cols[['Experience_Years']])

display(encoded_data[['age_minmax', 'age_standard', 'E_MinMax', 'E_standard']].head())
```
